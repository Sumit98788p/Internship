# Question 1: Scrape the details of most viewed videos on YouTube from Wikipedia. Url = https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos You need to find following details:

```python
import selenium
from selenium import webdriver
import pandas as pd
from selenium.webdriver.common.by import By
import warnings
warnings.filterwarnings("ignore")
import time
from selenium.common.exceptions import NoSuchElementException 
from selenium.common.exceptions import ElementClickInterceptedException
```

```python
driver = webdriver.Chrome()
```

```python
driver.get("https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos")
```

```python
rank =[]
name =[]
artist =[]
upload = []
views =[]

Rank  = driver.find_elements(By.XPATH,'//table[@class="wikitable sortable jquery-tablesorter"]//tbody//tr//td[1]')
for i in Rank:
    rank.append(i.text)
Name  = driver.find_elements(By.XPATH,'//table[@class="wikitable sortable jquery-tablesorter"]//tbody//tr//td[2]')
for i in Name:
    name.append(i.text)
Artists = driver.find_elements(By.XPATH,'//table[@class="wikitable sortable jquery-tablesorter"]//tbody//tr//td[3]')
for i in Artists:
    artist.append(i.text)
Upload_date = driver.find_elements(By.XPATH,'//table[@class="wikitable sortable jquery-tablesorter"]//tbody//tr//td[5]')
for i in Upload_date:
    upload.append(i.text)
Views = driver.find_elements(By.XPATH,'//table[@class="wikitable sortable jquery-tablesorter"]//tbody//tr//td[4]')
for i in Views:
    views.append(i.text)
```

```python
df = pd.DataFrame({'RANK':rank,'Video Name':name,'Artist Name':artist,'Upload date':upload,'Views in Billion':views})
df.head(30)
```

```python
driver.close()
```

# Question 4: Scrape the details of trending repositories on Github.com. Url = https://github.com/ 

```python
driver = webdriver.Chrome()
driver.get("https://github.com/")
```

```python
try:
    source = driver.find_element(By.XPATH,'//ul[@class="d-lg-flex list-style-none"]//li[3]//button[1]')
    source.click()
    trending = driver.find_element(By.XPATH,'//li[@class="HeaderMenu-item position-relative flex-wrap flex-justify-between flex-items-center d-block d-lg-flex flex-lg-nowrap flex-lg-items-center js-details-container js-header-menu-item"][3]//div//div[3]//ul//li[2]')
    trending.click()
except NoSuchElementException as e:

    print("Exception Raised: ", e)
```

```python
title =[]
description=[]
count = []
language =[]

try:
    Title = driver.find_elements(By.XPATH,'//article[@class="Box-row"]//h2//a')
    for i in Title:
        title.append(i.text)
except NoSuchElementException:
    title.append('-')
try:
    Descri = driver.find_elements(By.XPATH,'//article[@class="Box-row"]//p')
    for i in Descri:
        description.append(i.text)        
except NoSuchElementException:
    description.append('-')
    
try:
    Count  = driver.find_elements(By.XPATH,'//article[@class="Box-row"]//div[2]//a[2]')
    for i in Count:
        count.append(i.text)
except NoSuchElementException:
    count.append('-')

try:
    Lang_used = driver.find_elements(By.XPATH,'//article[@class="Box-row"]//div[2]//span[1]//span[2]')
    for i in Lang_used:
        language.append(i.text)
except NoSuchElementException:
    language.append('-')    
    
```

```python
print(len(title),len(description),len(count),len(language))
```

```python
df = pd.DataFrame({'Title':title})
df1 = pd.DataFrame({'Description':description})
df2 = pd.DataFrame({'Contributors Count':count})
df3 = pd.DataFrame({'Language Used':language})
df4 = pd.concat([df,df1,df2,df3],axis=1)
df4.head(25)
```

```python
name = []
Type=[]
task=[]
f_type=[]
instances=[]
attribute =[]
year =[]
Name = driver.find_elements(By.XPATH,'//div[@class="rounded-box bg-base-100"]//div//div[2]//h2//a')
for i in Name:
    name.append(i.text)
data = driver.find_elements(By.XPATH,'//div[@class="rounded-box bg-base-100"]//div//div[2]//div//div[2]//span')
for i in data:
    Type.append(i.text)
Task = driver.find_elements(By.XPATH,'//div[@class="rounded-box bg-base-100"]//div//div[2]//div//div[1]//span')
for i in Task:
    task.append(i.text)
feature= driver.find_elements(By.XPATH,'//div[@class="grid grid-cols-8 overflow-x-auto"]//table//tbody//tr//td[2]')
for i in feature:
    f_type.append(i)
Instance = driver.find_elements(By.XPATH,'//div[@class="rounded-box bg-base-100"]//div//div[2]//div//div[3]//span')
for i in Instance:
    instances.append(i.text)
attri = driver.find_elements(By.XPATH,'//div[@class="rounded-box bg-base-100"]//div//div[2]//div//div[4]//span')
for i in attri:
    attribute.append(i.text)
Year = driver.find_elements(By.XPATH,'//div[@class="grid grid-cols-8 overflow-x-auto"]//table//tbody//tr//td[3]')
for i in Year:
    year.append(i.text)

    
```

```python
print(len(name),len(Type),len(task),len(f_type),len(instances),len(attribute),len(year))
```
